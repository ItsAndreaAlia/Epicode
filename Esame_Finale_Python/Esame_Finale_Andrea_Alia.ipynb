# 2. Si chiede di trovare, per ogni continente:
# a. il numero di casi fin dall'inizio della pandemia
# b. la percentuale rispetto al totale mondiale del numero di casi


```python
# Punto 'A'

# Poichè quello fornito è un DataSet ordinato e già 'pulito', attarvereso il metodo '.', si va a richiamare la colonna 'date' per prendere visione del primo e dell'ultimo caso di Covid_19 registrato. 

df.date
```

```python
# Per trovare il numero totale dei casi fin dall'inizio della pandemia si utilizzi il metodo .sum() applicato alla colonna 'new_cases'

somma = df['new_cases'].sum()  

# poichè la somma ha come output un dato float con lo 0 dopo la virgola, lo si trasforma in intero, riassegnando la variabile
somma_int = int(somma)  

#si stampa il risultato
print(f"Il numero totale dei casi è:", {somma_int})
```

```python
# Si usa il metodo .unique() applicato alla colonnna 'continent' per verificare se i dati del nostro DataFrame corrisponodno a quelli che ci aspettiamo

df['continent'].unique()
```

```python
# attraverso il metodo .groupby() applicato alla colonna 'continent' si procede a calcolare la somma per ogni continente
totale_casi_per_continente = df.groupby('continent', dropna = False)['new_cases'].sum()
totale_casi_per_continente
```

```python
# punto B

# Andiamo ad estrarre tutti i valori dei totali dei casi per ogni continente

totale_casi_europa = totale_casi_per_continente['Europe'] # totale casi dell' Europa
print(f"Il numero totale dei casi dell'Europa è: {totale_casi_europa}")

totale_casi_asia = totale_casi_per_continente['Asia'] # totale casi dell' Asia
print(f"Il numero totale dei casi dell'Asia è: {totale_casi_asia}")

totale_casi_africa = totale_casi_per_continente['Africa'] # totale casi dell' Africa
print(f"Il numero totale dei casi dell'Africa è: {totale_casi_africa}")

totale_casi_north_america = totale_casi_per_continente['North America'] # totale casi del North America
print(f"Il numero totale dei casi del North America è: {totale_casi_north_america}")

totale_casi_south_america = totale_casi_per_continente['South America'] # totale casi del South America
print(f"Il numero totale dei casi del South America è: {totale_casi_south_america}")

totale_casi_oceania = totale_casi_per_continente['Oceania'] # totale casi dell'Oceania
print(f"Il numero totale dei casi dell'Oceania è: {totale_casi_oceania}")
```

```python
# Si calcoli la percentuale dei casi per ogni continente rispetto al totale

percentuale_europa = (totale_casi_europa / somma_int) * 100
print (f"La percentuale dei casi dell'Europa rispetto al totale è: {percentuale_europa}%")

percentuale_asia = (totale_casi_asia / somma_int) * 100
print (f"La percentuale dei casi dell'Asia rispetto al totale è: {percentuale_asia}%")

percentuale_africa = (totale_casi_africa / somma_int) * 100
print (f"La percentuale dei casi dell'Africa rispetto al totale è: {percentuale_africa}%")

percentuale_north_america = (totale_casi_north_america / somma_int) * 100
print (f"La percentuale dei casi del North America rispetto al totale è: {percentuale_north_america}%")

percentuale_south_america = (totale_casi_south_america / somma_int) * 100
print (f"La percentuale dei casi del South America rispetto al totale è: {percentuale_south_america}%")

percentuale_oceania = (totale_casi_oceania / somma_int) * 100
print (f"La percentuale dei casi dell'Oceania rispetto al totale è: {percentuale_oceania}%")
```

# 3. Selezionare i dati relativi all'Italia nel 2022 e mostrare con dei grafici adeguati:
# a. l'evoluzione del casi totali dall'inizio alla fine dell'anno
# b. il numero di nuovi casi rispetto alla data
# k. poiché i nuovi casi vengono registrati settimanalmente, filtrare via i dati che non hanno misurazioni

```python
# Punto A

# Procediamo per sicurezza a convertire la colonna 'data', nel formato proprio delle date
df['date'] = pd.to_datetime(df['date'])
```

```python
italia_2022 = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022)] 
italia_2022
```

```python
# andiamo ad ordinare (per sicurezza) i dati secondo il parametro 'data', riassegnando la variabile

italia_2022 = italia_2022.sort_values(by='date')
```

```python
# Creiamo il grafico impostando i nostri parametri desiderati

italia_2022.plot(x='date', y='total_cases', kind='line', \
                 figsize=(10, 6), grid=True, # ho inserito il parametro grid = true perchè ritengo fornisca una maggiore chiarezza per la visione
                 title="L'evoluzione del casi totali nel corso del 2022") 
```

```python
# Punto B

# si crei un grafico a barre per evidenziare i nuovi casi rispetto alla data
italia_2022.plot(x='date', y='new_cases', kind='bar', title="Numero di nuovi casi rispetto alla data")

# come si può ben notare, questo tipo di grafico vede una sovrapposizione di ddati sull'asse delle x, perciò andiamo a risolvere nella cella successiva
```

```python
# Per evitare la sovrapposizione di cui sopra, andiamo a filtrare i nostri dati mensilmente e a creare la colonna 'month' convertendo la
# data attraverso il metodo .dt.to_period ed usando uno degli alias del periodo ( nel nostro caso 'M')

italia_2022['month'] = italia_2022['date'].dt.to_period('M')
```

```python
# attarverso la funzione group.by)() andiamo a raggruppare per mese e a sommare le colonne
mesi_italia_2022 = italia_2022.groupby('month')['new_cases'].sum()

# Crea il grafico a barre per i dati mensili
mesi_italia_2022.plot(kind='bar', figsize=(10, 6), grid=True, title="Nuovi casi mensili in Italia nel 2022")
```

```python
# punto k

# poiché i nuovi casi vengono registrati settimanalmente, è possibile filtrare via i dati che non hanno misurazioni ponendo i dati della colonna 'new_cases' > 0
italia_2022_maggiore_0 = italia_2022 = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2022) & (df['new_cases'] > 0)]
italia_2022_maggiore_0
```

```python
# si vada a creare il grafico relativo al filtro appena creato

italia_2022_maggiore_0.plot(x='date', y='new_cases', kind='bar', figsize=(10, 6),
                 title="Nuovi casi settimanali in Italia (nuovi casi > 0)")
```

# 4. Riguardo le nazioni di Italia, Germania e Francia mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU) da maggio 2022 (incluso) ad aprile 2023 (incluso) a. scrivere un breve commento a riguardo¶


```python
# si vada innanzitutto a creare il filtro per le nazioni e il periodo temporale richiesto

filtro_terapia_intensiva = (df['location'].isin(['Italy', 'Germany', 'France'])) & (df['date'] >= '2022-05-01') \
                            & (df['date'] <= '2023-04-30')  & (df['icu_patients'] > 0)
terapia_intensiva = df[filtro_terapia_intensiva]
```

```python
# si crei il boxplot
terapia_intensiva.boxplot(column='icu_patients', by='location', grid=True, figsize = (10,6))
```

```python
# Punto A - Commento

# Come si può ben notare, la Francia e la Germania constano di un numero di pazienti in terapia intensiva decisamnete maggiore rispetto all'Italia.
# Nella Germania si evidenzia una grande dispersione rispetto alla Francia, quindi è possibile evincere come i dati abbiano un indice di variabilità maggiore rispetto alla regione francese.
# L'italia, d'altra parte, si caratterizza per un numero di pazienti significativamente inferiore, godendo di una distribuzione che si rivela più uniforme.
```

# 5. Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2023:
# a. mostrare la somma dei pazienti ospitalizzati per ognuna
# b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno¶

```python
# Punto 1
# si creino dei filtri per ogni paese che indicano i casi dei pazienti ospitalizzati nel 2023

italia = df[(df['location'] == 'Italy') & (df['date'].dt.year == 2023)] 
francia = df[(df['location'] == 'France') & (df['date'].dt.year == 2023)]
germania = df[(df['location'] == 'Germany') & (df['date'].dt.year == 2023)]
spagna = df[(df['location'] == 'Spain') & (df['date'].dt.year == 2023)]
```

```python
# si utilizzano i filtri per andare a calcolare la somma dei pazienti ospitalizzati per ognuna nazione

ospitalizzati_italia = italia['hosp_patients'].sum()
print(f"La somma dei pazienti ospitalizzati in Italia nel 2023: {ospitalizzati_italia}")

ospitalizzati_francia = francia['hosp_patients'].sum()
print(f"La somma dei pazienti ospitalizzati in Francia nel 2023: {ospitalizzati_francia}")

ospitalizzati_germania = germania['hosp_patients'].sum()
print(f"La somma dei pazienti ospitalizzati in Germania nel 2023: {ospitalizzati_germania}")

ospitalizzati_spagna = spagna['hosp_patients'].sum()
print(f"La somma dei pazienti ospitalizzati in Spagna nel 2023: {ospitalizzati_spagna}")
```

```python
# Punto 2

# Effettuare un'indagine che va a verificare la presenza di dati nulli

dati_nulli_italia = italia[italia['hosp_patients'].isna()] #verifica dati nulli per l'Italia
dati_nulli_italia

# Output: In Italia non risultano dati nulli, pertanto la problematica non si pone
```

```python
# si svolga un analogo procedimento per i paesi interessati

dati_nulli_germania = germania[germania['hosp_patients'].isna()]
dati_nulli_germania

# Output: La Germania conta 0 pazienti ospitalizzati nel 2023, nonchè 365 dati nulli. Pertanto potrebbe essere presa in considerazione l'idea di escludere la Germania dall'analisi
```

```python
dati_nulli_spagna = spagna[spagna['hosp_patients'].isna()]
dati_nulli_spagna

# Output: La Spagna consta di 203 dati nulli su un totale di casi ospitalizzati largamente superiore, ragion per cui è possibile considerare l'idea di sostituire i dati nulli con la mediana
```

```python
dati_nulli_francia = francia[francia['hosp_patients'].isna()]
dati_nulli_francia

# Output: Analogamente alla Spagna, la Francia vede un numero nettamente inferiore di dati ulli risppetto al totale dei casi, per cui è accettabile l'idea di applicare la soluzione precedente anche alla suddetta fattispecie
```
